## 여러 데이터 비교를 위한 결과 HTMl 수정

from IPython.display import display_html
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str += df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'), raw=True)

## (1~4번) 기준 두개의 DataFrame 데이터 준비

```python
import pandas as pd

df1 = pd.DataFrame({
    '사원번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['John', 'Park', 'Han', 'Kim', 'Choi', 'Back', 'Denny']
}, columns=['사원번호', '이름'])

df2 = pd.DataFrame({
    '사원번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '시급': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['사원번호', '시급'])

print("※ 기준 두 데이터프레임 데이터 ")
display_side_by_side(df1, df2)
```

### 1) merge 함수 inner join

```python
display_side_by_side(pd.merge(df1,df2), pd.merge(df1, df2, on="사원번호", how="inner"))
```

### 2) merge 함수 outer join

```python
display_side_by_side(df1,df2, pd.merge(df1,df2, how="outer"))
```

### 3) merge 함수 left join

```python
display_side_by_side(df1,df2, pd.merge(df1,df2, how="left"))
```

### 4) merge 함수 right join

```python
display_side_by_side(df1,df2, pd.merge(df1,df2, how="right"))
```

## (1~2번) 기준 두개의 DataFrame 데이터 준비

```python
import pandas as pd

df1 = pd.DataFrame({
    '사원번호': [1001, 1001, 1003, 1004, 1005, 1006, 1007],
    '이름': ['John', 'Park', 'Han', 'Kim', 'Choi', 'Back', 'Denny']
}, columns=['사원번호', '이름'])

df2 = pd.DataFrame({
    '사원넘버': [1001, 1001, 1005, 1006, 1008, 1001],
    '시급': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['사원넘버', '시급'])

print("※ 기준 두 데이터프레임 데이터 ")
display_side_by_side(df1, df2)
```

### 1) merge 함수 left_on, right_on 

```python
display_side_by_side(df1, df2, pd.merge(df1, df2, left_on="사원번호", right_on="사원넘버"))
```

### 2) merge 함수 left_index, right_index

```python
df1.set_index('사원번호', inplace=True)
df2.set_index('사원넘버', inplace=True)
```

```python
display_side_by_side(df1,df2,pd.merge(df1,
                                      df2, 
                                      left_index=True,
                                      right_index=True))
```

# join 함수

```python
display_side_by_side(df1,df2,df1.join(df2, how="inner"))
```

# concat 함수

```python
# index 열 (사원번호, 사원넘버)에서 해체
df1.reset_index(inplace=True)
df2.reset_index(inplace=True)
```

### 행으로 데이터 연결

```python
display_side_by_side(df1,df2,pd.concat([df1,df2], axis=0))
```

### 열로 데이터 연결

```python
display_side_by_side(df1,df2,pd.concat([df1,df2], axis=1))
```
