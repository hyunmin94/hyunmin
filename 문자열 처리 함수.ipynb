```python
# 전체 함수에 사용할 변수(split, strip 제외)
str_data = "Python is the best choice!"
# strip 함수 변수
strip_test_data = '    Python     '
# split 함수 변수
split_test_data = 'a,b,c,d,e,f'

# count 함수 [문자 'i'의 갯수를 반환 ]
print('※ count 함수')
print(str_data.count('i'), end='\n\n')

# find 함수 [문자 'i'의 처음 나온 위치를 반환]
print('※ find 함수')
print(str_data.find('m'), end='\n\n')

# index 함수 [문자 'i'의 처음 나온 위치를 반환]
print('※ index 함수')
print(str_data.index('n'), end='\n\n')

# join 함수 [a.join(b) 일경우 b의 문자열을 각각 a 문자를 구분자로 연결]
# 같은 결과 - > print(','.join(['a','b','c','d','e']))
print('※ join 함수')
print(','.join('abcde'), end='\n\n')

# upper 함수 [소문자를 대문자로 치환]
print('※ upper 함수')
print(str_data.upper(), end='\n\n')

# lower 함수 [대문자를 소문자로 치환]
print('※ lower 함수')
print(str_data.lower(), end='\n\n')

# lstrip 함수 [왼쪽 공백 제거]
print('※ lstrip 함수')
print(strip_test_data.lstrip(), end='\n\n')

# rstrip 함수 [오른쪽 공백 제거]
print('※ rstrip 함수')
print(strip_test_data.rstrip(), end='\n\n')

# strip 함수 [양쪽 공백 제거]
print('※ strip 함수')
print(strip_test_data.strip(), end='\n\n')

# replace 함수 [문자열 바꾸기]
print('※ replace 함수')
print(str_data.replace('Python','Java'), end='\n\n')

# split 함수 [해당 구분자로 문자열을 나눔으로써 리스트형태로 반환]
print('※ split 함수')
print(split_test_data.split(','), end='\n\n')
```
