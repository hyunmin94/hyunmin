```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# 그래프에 한글 출력시 폰트 변경 (기본폰트 : sans-serif)
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
# 그래프에서 마이너스(-)폰트 깨짐 방지
matplotlib.rcParams['axes.unicode_minus'] = False

names = ['철수','지훈','명환','상준','동훈','우현','치국']                    
weight = [50,80,65,80,60,100,35]
height = [160,180,178,165,179,190,150]
# 마커 사이즈 (몸무게 비율) 
size = np.array(weight) * 10
# 마커 컬러 
colors = ['b','k','y','c','r','g','m']

plt.scatter(x = height, y = weight, s = size, c = colors, marker= 'd', alpha=0.5)
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height & Weight')

# 마커 위치에 해당 인원 이름 출력
for height,weight,name in zip(height,weight,names) :
    plt.text(height,weight,name)
plt.show()
```

```python
import matplotlib
import numpy as np
import pandas as pd

# 그래프에 한글 출력시 폰트 변경 (기본폰트 : sans-serif)
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
# 그래프에서 마이너스(-)폰트 깨짐 방지
matplotlib.rcParams['axes.unicode_minus'] = False

names = ['철수','지훈','명환','상준','동훈','우현','치국']                    
weight = [50, 80, 65, 80, 60, 100, 35]
height = [160, 180, 178, 165, 179, 190, 150]
# 마커 사이즈 (몸무게 비율) 
size = np.array(weight) * 10
# 마커 컬러 
colors = ['b','k','y','c','r','g','m']
# DataFrame 객체 생성 
dataFrame_obj = pd.DataFrame({'Weight': weight, 'Height': height})

scatter_plot = dataFrame_obj.plot.scatter(x = 'Height', y = 'Weight', s = size, c = colors, marker= 'd', alpha=0.5)
# 같은 그래프 생성
# ▶ scatter_plot = dataFrame_obj.plot(kind='scatter',x = 'Height', y = 'Weight', s = size, c = colors, marker= 'd', alpha=0.5)

# x축 라벨
scatter_plot.set_xlabel('Height(cm)')
# y축 라벨
scatter_plot.set_ylabel('Weight(kg)')
# 타이틀
scatter_plot.set_title('Height & Weight')

# 마커 위치에 해당 인원 이름 출력
for height,weight,name in zip(height,weight,names) :
    plt.text(height,weight,name)
plt.show()
```
