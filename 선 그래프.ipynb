```python
import matplotlib.pyplot as plt
import numpy as np

legend_labels = ['A','B','C','D']

x = np.arange(0,4,0.5)

y1 = x + 1
y2 = x + 3
y3 = x + 5
y4 = x + 7

# 스타일: 선 컬러, 선 스타일, 마커스타일 (순서 무관/생략 가능) 
plt.plot(x,y1,'r-.>',x,y2,'b--<',x,y3,'c:o',x,y4,'k-d')
# 범례 생성 및 위치 지정
plt.legend(legend_labels, loc='upper left')
# X축 라벨
plt.xlabel('X-axis')
# Y축 라벨
plt.ylabel('Y-axis')
# 타이틀(title)
plt.title('Graph Title')
# 그리드(Grid)지정 [True/False] (기본값: True)
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

legend_labels = ['A','B','C','D']

x = np.arange(0,4,0.5)

y1 = x + 1
y2 = x + 3
y3 = x + 5
y4 = x + 7

# DataFrame 객체 생성
data_frame = pd.DataFrame({'A':y1,'B':y2,'C':y3,'D':y4}, index = x, columns=legend_labels)

# 스타일: 선 컬러, 선 스타일, 마커스타일 (순서 무관/생략 가능) 
math_plot = data_frame.plot(grid=True, style =['r-.>','b--<','c:o','k-d'], title='Graph Title')

# 같은 그래프 생성
# ▶ math_plot = data_frame.plot(kind='line', grid=True, style =['r-.>','b--<','c:o','k-d'], title='Graph Title')

# ▶ math_plot = data_frame.plot.line(grid=True, style =['r-.>','b--<','c:o','k-d'], title='Graph Title')

# X축 라벨
math_plot.set_xlabel('X-axis')
# Y축 라벨
math_plot.set_ylabel('Y-axis')

# 타이틀(title)  - > plot() 인자로 설정
# math_plot.set_title('Graph Title')

# 그리드(Grid)지정 - > plot() 인자로 설정
# plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

legend_labels = ['A','B']

x = np.arange(0,4,0.5)

y1 = x + 1
y2 = x + 3

# DataFrame 객체 생성
data_frame2 = pd.DataFrame({'A':y1,'B':y2}, index = x, columns=legend_labels)

# X축 라벨 : x 값으로 자동 설정(data_frame2 컬럼명)
# 범례 : y 값으로 자동 설정 (data_frame2 컬럼명) 
# x,y는 data_frame의 컬럼명으로 지정하여 해당 열의 데이터를 가져온다
math_plot = data_frame2.plot(x= 'A', y = 'B', style =['r-.>'])

# 타이틀(title)
math_plot.set_title('Graph Title')
# 그리드(Grid)
plt.grid()
```
